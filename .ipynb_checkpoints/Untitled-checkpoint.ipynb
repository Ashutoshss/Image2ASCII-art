```python
import cv2 as cv
import numpy as np
import os

cwd = os.getcwd()
path = os.path.join(cwd,'image_file','letter-a-letter-alphabet-mouse-pad.jpg')
```

```python
def img2ASCII(image, size):
    # Define the ASCII intensity characters
    ascii_intensity = " .-=+:cba!?0123456789$W#@Ã‘"
    
    # Initialize the result string
    ascii_string = ""

    # Loop over every pixel in the grayscale image
    for y in range(size[0]):  # height
        for x in range(size[1]):  # width
            # Get pixel value
            pixel_value = image[y, x]
            # Map the pixel value to the corresponding character
            ascii_char = ascii_intensity[pixel_value * (len(ascii_intensity) - 1)  // 255]
            ascii_string += ascii_char  # Add the character to the ascii_string
        ascii_string += "\n"  # New line at the end of each row

    return ascii_string  # Return the complete ASCII string
```

```python
image = cv.imread(path)
grayScale = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
resize_ = cv.resize(grayScale,(200,200),interpolation = cv.INTER_LINEAR)
size = resize_.shape
print(size)
result = img2ASCII(resize_,size)
print(result)
cv.imshow("Original Image",image)
cv.waitKey(0)
cv.destroyAllWindows()
```
