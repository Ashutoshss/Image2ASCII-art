```python
import cv2 as cv
import numpy as np
import os

cwd = os.getcwd()
path = os.path.join(cwd,'image_file','Untitled.jpeg')
```

```python
def img2ASCII(image, size):
    # ascii_intensity = " .-=+:cba!?0123456789$W#@Ñ"
    ascii_intensity = "Ñ@#W$9876543210?!abc:+=-. " 
    ascii_string = ""

    for y in range(size[0]):  # height
        for x in range(size[1]):  # width
            pixel_value = image[y, x]
            # Map the pixel value
            ascii_char = ascii_intensity[pixel_value * (len(ascii_intensity) - 1)  // 255]
            # print(pixel_value * (len(ascii_intensity) - 1)  // 255)
            ascii_string += ascii_char  
        ascii_string += "\n"  

    return ascii_string 
```

```python
image = cv.imread(path)
grayScale = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
resize_ = cv.resize(grayScale,(100,100),interpolation = cv.INTER_LINEAR)
size = resize_.shape
# print(size)
result = img2ASCII(resize_,size)
print(result)
text_file = os.path.join(cwd,'ascii_art.txt')

with open("text_file", "w") as file:
    file.write(result)
cv.imshow("Original Image",resize_)
cv.waitKey(0)
cv.destroyAllWindows()
```
